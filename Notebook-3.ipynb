```sql
-- Tvorba VIEW seskupení podle Country a Year.                                                              CREATE
CREATE VIEW CatastrophesGroupDataByCountryAndYear AS
SELECT c.Country, c.ISO, c.Start_Year, 
    [Direct_economic_loss_USD],
    COUNT(DisNo) AS CatastrophesCount
FROM [Catastrophes new] c
    LEFT JOIN [Direct_disaster_economic_loss_modified] dl ON c.ISO = dl.ISO AND c.Start_Year = dl.Year
GROUP BY c.Country, c.ISO, c.Start_Year, dl.[Direct_economic_loss_USD];
```

```sql
-- V jaký rok a kde byl nejvyšší počet katastrof?
SELECT Country, Start_Year, MAX(CatastrophesCount) AS MaxCount
FROM [dbo].[CatastrophesGroupDataByCountryAndYear]
GROUP BY Country, Start_Year
ORDER BY MaxCount DESC;
```
