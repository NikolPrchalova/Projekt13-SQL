### DIRECT\_DISASTER\_ECONOMIC\_LOSS

Pozor! jedná se pouze o přírodní katastrofy.

```sql
-- Kontrola relevance údajů. Zda v tabulce [Direct_disaster_economic_loss] za dané roky stoupá hodnota USD.         INFO!
-- Kdyby nestouapala, bylo by nutno ověřit, zda jsou hodnoty přepočteny (přes iflaci) k danému (jakému?) roku
WITH EconomicLossByYear AS (
    SELECT Year, SUM([Direct_economic_loss_USD]) AS Totalloss
    FROM [dbo].[Direct_disaster_economic_loss]
    GROUP BY Year
)

SELECT Year, Totalloss,
    (Totalloss - LAG(Totalloss) OVER (ORDER BY Year)) / LAG(Totalloss) OVER (ORDER BY Year) * 100 AS PercentChange
FROM EconomicLossByYear 
ORDER BY Year;
```

```sql
-- Vizuální posouzení hodnot v tabulce [Catastrophes new] a [Direct_disaster_economic_loss]                     INFO
-- Hodnoty za zemi a rok, které nejsou v nové tabulce Direct... doplnit? - doplněno níže
WITH TotalDamageByISOandYear AS (
    SELECT c.ISO, c.[Start_Year], SUM(c.[Total_Damage_000_US]) AS TotalDamage
    FROM [Catastrophes new] c
    WHERE [Disaster_Group] = N'Natural' AND [Total_Damage_000_US] IS NOT NULL
    GROUP BY c.ISO, c.[Start_Year]
)

SELECT c.ISO, c.[Start_Year], c.TotalDamage, d.ISO, d.Year, d.[Direct_economic_loss_USD]
FROM TotalDamageByISOandYear c
    FULL JOIN [Direct_disaster_economic_loss] d ON c.ISO = d.ISO AND c.[Start_Year] = d.Year
ORDER BY d.ISO, d.Year, c.ISO;
```

### Direct\_disaster\_economic\_loss\_modified

```sql
-- Nová tabulka "Direct_disaster_economic_loss_modified", kde jsou řádky z Direct_disaster_economic_loss a tam,     CREATE
-- kde chybí řádky (ISO, Year) jsou doplněny řádky z Catastrophes new. Tam kde je tak provedeno je v doplněném 
-- sloupci "Edit_cost" doplněna 1 (v nedoplněných řádcích je 0). 

WITH TotalDamageByISOandYear AS (
    SELECT c.ISO, c.[Start_Year], SUM(c.[Total_Damage_000_US]) AS TotalDamage
    FROM [Catastrophes new] c
    WHERE [Disaster_Group] = N'Natural' AND [Total_Damage_000_US] IS NOT NULL
    GROUP BY c.ISO, c.[Start_Year]
)

-- Vytvoření nové tabulky "Direct_disaster_economic_loss_modified" s Edit_cost sloupcem
SELECT ISNULL(c.ISO, d.ISO) AS ISO, ISNULL(c.[Start_Year], d.Year) AS Year,
       CASE
           WHEN c.ISO IS NOT NULL AND c.[Start_Year] IS NOT NULL THEN
               CASE
                   WHEN d.[Direct_economic_loss_USD] IS NULL THEN c.TotalDamage -- * 1000 (opraveno o buňku níže)
                   ELSE d.[Direct_economic_loss_USD]
               END
           ELSE d.[Direct_economic_loss_USD]
       END AS Direct_economic_loss_USD,
       CASE
           WHEN (d.[Direct_economic_loss_USD] = 0 OR d.[Direct_economic_loss_USD] IS NULL) AND c.TotalDamage IS NOT NULL THEN 1
           ELSE 0
       END AS Edit_cost
INTO Direct_disaster_economic_loss_modified
FROM TotalDamageByISOandYear c
FULL JOIN [Direct_disaster_economic_loss] d ON c.ISO = d.ISO AND c.[Start_Year] = d.Year
ORDER BY ISO, Year;

-- EDIT_COST
-- V tabulce Direct_disaster_economic_loss_modified se vyskytuje sloupec Edit_cost, který nám zohledňuje,
-- z které tabulky je hodnota Direct_economic_loss_USD převzata. Pokud je 1, je ztráta převzata z původního
-- datasetu catastrophes new ze sloupce Total Damage, pokud je 0 - je ponecháný dataset zohledňující přímo
-- ekonomické ztráty katastrof (direct-disaster-economic-loss).
```

```sql
-- Hodnoty Total_Damage jsou v datasetu v tis. USD - zapomněla jsem to zohlednit.                                CREATE
UPDATE Direct_disaster_economic_loss_modified
SET Direct_economic_loss_USD = CASE
    WHEN Edit_cost = 1 THEN Direct_economic_loss_USD * 1000
    ELSE Direct_economic_loss_USD
END;
```

```sql
-- Smazání všech řádku, které v Direct_economic_loss_USD obsahují 0.                                             CREATE
-- (nepotřebuji řádky, které nemají žádnou vypovídající hodnotu)                                          
DELETE FROM Direct_disaster_economic_loss_modified
WHERE Direct_economic_loss_USD = 0;
```

Sloupec 'Ratio\_economic\_loss\_to\_GDP'

```sql
-- Přidání prázdného sloupce Ratio_economic_loss_to_GDP.                                                        CREATE
ALTER TABLE Direct_disaster_economic_loss_modified
ADD Ratio_economic_loss_to_GDP FLOAT; 
```

```sql
-- Naplnění prázdného sloupce podílem: economic loss / GDP.                                                      CREATE
UPDATE dl
SET Ratio_economic_loss_to_GDP = ROUND(dl.Direct_economic_loss_USD / gdp.GDP, 4)
FROM Direct_disaster_economic_loss_modified dl
    JOIN Gross_Domestic_Product gdp ON dl.ID = gdp.ID;
```

```sql
-- Protože se smazali některé řádky v GDP a nepropsalo se to do Ratio_economic_loss_to_GDP,                         CREATE
-- tak vše smazat a spustit naplnění sloupce Ratio.. znovu.
UPDATE [Direct_disaster_economic_loss_modified]
SET Ratio_economic_loss_to_GDP = NULL;
```

### Validující a dopňující informace do článku z tabulky Direct\_disaster\_economic\_loss\_modified

```sql
-- V jakých letech jaké země nemají podíl k HDP - validace k provedeným grafům                                          INFO
SELECT DISTINCT(ISO), Year, Ratio_economic_loss_to_GDP
FROM Direct_disaster_economic_loss_modified 
WHERE YEAR >= 2000 AND YEAR < 2023 AND Ratio_economic_loss_to_GDP IS NULL;
```

```sql
-- Statistické veličiny
SELECT COUNT(Ratio_economic_loss_to_GDP) as CountRatio, 
       MIN(Ratio_economic_loss_to_GDP) as MinRatio, 
       MAX(Ratio_economic_loss_to_GDP) as MaxRatio, 
       VAR(Ratio_economic_loss_to_GDP)  as Rozptyl,
       STDEV(Ratio_economic_loss_to_GDP) as SměrodatnáOdchylka, -- Odchylka od průměru
       AVG(Ratio_economic_loss_to_GDP) as AvgRatio
FROM [dbo].[Direct_disaster_economic_loss_modified]
WHERE YEAR >= 2000 AND YEAR < 2023 AND Ratio_economic_loss_to_GDP IS NOT NULL

-- MEDIAN
SELECT DISTINCT
  PERCENTILE_CONT(0.5) WITHIN GROUP (ORDER BY Ratio_economic_loss_to_GDP) OVER () AS Median
FROM [dbo].[Direct_disaster_economic_loss_modified]
WHERE YEAR >= 2000 AND YEAR < 2023 AND Ratio_economic_loss_to_GDP IS NOT NULL;
```

```sql
-- 10 %
-- Jedinečné země, které přesáhly v daných letech hranici > 10 %:                                                                                                               INFO                                                                                                            INFO
SELECT DISTINCT(ISO), g.Country_Name, SUM(Ratio_economic_loss_to_GDP) AS TotalRatio
FROM Direct_disaster_economic_loss_modified dl
    JOIN Gross_Domestic_Product g ON dl.ID = g.ID
WHERE dl.YEAR >= 2000 AND dl.YEAR < 2023
GROUP BY dl.ISO, g.Country_Name
HAVING SUM(Ratio_economic_loss_to_GDP) > 0.1
ORDER BY TotalRatio DESC;


-- země v daných letech přesáhly hranici Ratio > 10 % a kolikrát:
SELECT
    ISO,
    Country_Name,
    COUNT(*) AS Exceeds_Threshold_Count
FROM
    (
        SELECT
            dl.ISO,
            g.Country_Name,
            SUM(dl.Ratio_economic_loss_to_GDP) AS TotalRatio
        FROM
            Direct_disaster_economic_loss_modified dl
            JOIN Gross_Domestic_Product g ON dl.ID = g.ID
        WHERE
            dl.YEAR >= 2000 AND dl.YEAR < 2023
        GROUP BY
            dl.ISO, g.Country_Name
        HAVING
            SUM(dl.Ratio_economic_loss_to_GDP) > 0.1
    ) AS Subquery
GROUP BY
    ISO, Country_Name
ORDER BY
    Exceeds_Threshold_Count DESC;

-- 56 zemí přesáhlo v SUMĚ Ratio > 10 %
-- x = 100*56 / 1949 = 0,82 % -> Tyto výskyty tvoří 0,82 % všech záznamů.
```
